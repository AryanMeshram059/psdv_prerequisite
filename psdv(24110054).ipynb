```python
#QUE1]=>

#creating a list "l" of first 20 natural numbers
l=[x for x in range(1,21)]
print(f"List of first 20 natural numbers: {l}")

#CALCULATING THE SUM OF ALL EVEN NUMBERS OF THE LIST 
evensum=0
for i in l:
    if i%2==0:
        evensum+=i
print(f"sum of all even numbers: {evensum}")

# CALCULATING THE PRODUCT OF ODD NUMBERS IN THE LIST
oddproduct=1
for i in l:
    if i%2!=0:
        oddproduct*=i
print(f"Product of all odd numbers of the list: {oddproduct}")

# to calculate mean
mean=sum(l)/len(l)
print(f"mean: {mean}")

# TO CALCULATE MEDIAN 
l.sort()
n=len(l)
if n%2==0:
    num1=l[n//2]
    num2=l[(n//2)-1]
    med=(num1+num2)/2
else:
    med=l[(n//2)]
print(f"median: {med}")
```

```python
#QUE3]=>

# CREATING A FUNCTION "f" THAT RETURNS THE FIBONACCI NUMBER OF n'th TERM:
def f(n):
    if n==0:
        return(0)
    elif n==1:
        return(1)
    else:
        return(f(n-1)+f(n-2))

#TAKING INPUT "n" :
n=int(input())

#CREATING LIST OF FIBONACCI NUMBERS TILL n'th TERM:
l=[]
for i in range(n+1):
    l.append(f(i))

#PRINTING THE LIST:
print(l)

# PLOTTING FIBONACCI SEQUENCE TILL n'th TERM USING A LINE PLOT :
from matplotlib import pyplot as plt
x=[i for i in range(n+1)]
y=l
plt.ylabel("fibonacci numbers")
plt.xlabel("terms")
plt.plot(x,y)
plt.show()
```

```python
#QUE6]=>

def isPalindrome(sen):
    new_sen="".join(sen.split()).lower()
    rever_sen=new_sen[::-1]
    if new_sen==rever_sen:
        return(True)
    else:
        return(False)

sen=input()
print(isPalindrome(sen))
```

```python
#QUE12]=>

from matplotlib import pyplot as plt
l=eval(input())
longest=l[0]

# to find the largest string: 
for i in l:
    if len(i)>=len(longest):
        longest=i
print(f"longest string: {longest}")

#counting the number of string starting with a vowel:
count=0
vowels=["a","e","i","o","u"]
for i in l:
    if i[0].lower() in vowels:
        count+=1
print(f"Number of string starting with vowels: {count}")

# ploting the histogram of string lengths:
lengths=[len(x) for x in l]
plt.hist(lengths,color="yellow",edgecolor="orange",linewidth=4)
plt.xlabel("string length")
plt.ylabel("frequency")
plt.show()
```

QUE 14: Word Filtering and Sorting

```python
from matplotlib import pyplot as plt
l=input().strip(". ,").lower().split()
small=[]
for i in l :
    #print(f"i: {i}")
    if len(i)<4:
        small.append(i)
    #print(f"l: {l}")
for i in small:
    l.remove(i)
l.sort()
print(f"sorted list: {l}")
lengths=[len(i) for i  in l]
plt.bar(l,lengths,color="skyblue",edgecolor="blue",linewidth=3)
plt.xlabel("words")
plt.ylabel("length")
plt.show()
```

QUE 15: ANAGRAM FINDER

```python
word=input()
words=eval(input())
temp=[i for i in word]
temp.sort()
data=[temp]
output=[]
for i in words:
    letters=[j for j in i.lower()]
    letters.sort()
    if letters in data:
        output.append(i)
    else:
        data.append(letters)
print(output)
```
