# QUE 1: List Operations and Basic Statistics

````python
#creating a list "l" of first 20 natural numbers
l=[x for x in range(1,21)]
print(f"List of first 20 natural numbers: {l}")

#CALCULATING THE SUM OF ALL EVEN NUMBERS OF THE LIST 
evensum=0
for i in l:
    if i%2==0:
        evensum+=i
print(f"sum of all even numbers: {evensum}")

# CALCULATING THE PRODUCT OF ODD NUMBERS IN THE LIST
oddproduct=1
for i in l:
    if i%2!=0:
        oddproduct*=i
print(f"Product of all odd numbers of the list: {oddproduct}")

# to calculate mean
mean=sum(l)/len(l)
print(f"mean: {mean}")

# TO CALCULATE MEDIAN 
l.sort()
n=len(l)
if n%2==0:
    num1=l[n//2]
    num2=l[(n//2)-1]
    med=(num1+num2)/2
else:
    med=l[(n//2)]
print(f"median: {med}"),
````

# QUE 2: Mutable vs. Immutable Data Types

## Difference Between Mutable and Immutable Objects in Python

In Python, **objects** can be categorized as either **mutable** or **immutable** based on whether their value can be changed after they are created.

---

## Mutable Objects
- **Definition**: Mutable objects are objects whose values can be modified after creation.
- **Examples**: 
  - `list`
  - `dict`
  - `set`
- **Behavior**: You can change the contents of a mutable object without changing its identity (i.e., the object remains the same in memory).

### Example:
```python
my_list = [1, 2, 3]
print(my_list)  # Output: [1, 2, 3]

# Modifying the list
my_list.append(4)
print(my_list)  # Output: [1, 2, 3, 4]


# QUE 3: Fibonacci Sequence

````python
# CREATING A FUNCTION "f" THAT RETURNS THE FIBONACCI NUMBER OF n'th TERM:
def f(n):
    if n==0:
        return(0)
    elif n==1:
        return(1)
    else:
        return(f(n-1)+f(n-2))

#TAKING INPUT "n" :
n=int(input())

#CREATING LIST OF FIBONACCI NUMBERS TILL n'th TERM:
l=[]
for i in range(n+1):
    l.append(f(i))

#PRINTING THE LIST:
print(l)

# PLOTTING FIBONACCI SEQUENCE TILL n'th TERM USING A LINE PLOT :
from matplotlib import pyplot as plt
x=[i for i in range(n+1)]
y=l
plt.ylabel("fibonacci numbers")
plt.xlabel("terms")
plt.plot(x,y)
plt.show()
````

# QUE 6: Palindrome Check

````python
def isPalindrome(sen):
    new_sen="".join(sen.split()).lower()
    rever_sen=new_sen[::-1]
    if new_sen==rever_sen:
        return(True)
    else:
        return(False)

sen=input()
print(isPalindrome(sen))
````

# QUE 9: Merging Dictionaries

````python
# declaring two dictionaries:
dict1={"a":10,"b":20}
dict2={"b":5,"c":15}

# adding the two dictionaries:
c={i:dict1[i] for i in dict1}
for i in dict2:
    if i in c:
        c[i]+=dict2[i]
    else:
        c[i]=dict2[i]

# printing the result:
print(c)
````

# QUE 11: Remove Duplicates:

# QUE 12: String Operations

````python
from matplotlib import pyplot as plt
l=eval(input())
longest=l[0]

# to find the largest string: 
for i in l:
    if len(i)>=len(longest):
        longest=i
print(f"longest string: {longest}")

#counting the number of string starting with a vowel:
count=0
vowels=["a","e","i","o","u"]
for i in l:
    if i[0].lower() in vowels:
        count+=1
print(f"Number of string starting with vowels: {count}")

# ploting the histogram of string lengths:
lengths=[len(x) for x in l]
plt.hist(lengths,color="yellow",edgecolor="orange",linewidth=4)
plt.xlabel("string length")
plt.ylabel("frequency")
plt.show()
````

# QUE 14: Word Filtering and Sorting

# QUE 18: Word Count and Frequency

````python
with open("file.txt","r+") as f:
    data={}
    count=0
    txt=f.readline().split()
    most=txt[0]
    for i in txt:
        count+=1
        if i in data:
            data[i]+=1
        else:
            data[i]=1
            if data[i]>data[most]:
                most=i
print(data)
print(f"word count: {count}")
print(f"most ocuuring word: {most}")
````

# QUE 22: Normalization

````python
from matplotlib import pyplot as plt
l=[12, 14, 32, 96, 7, 62, 88, 67, 25, 84, 10, 29, 10, 25, 68, 13, 58, 41, 44, 44]
norm_l=[]
for i in l:
    v=(i-min(l))/(max(l)-min(l))
    norm_l.append(v)
print("number==>normalised value")
for i in range(len(l)):
    print(f"{l[i]}==>{norm_l[i]}")
plt.plot(norm_l)
plt.show()
````

# QUE 25: Leap Year Checker

````python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

year = int(input("Enter a year: "))

if is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
````
